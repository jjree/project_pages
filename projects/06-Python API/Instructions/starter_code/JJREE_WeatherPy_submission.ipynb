----
1. The first observation involves the relationship between latitude and temperature. The results show that the temperature is positively correlated to the latitude as it approaches the equator (lat = 0) and negatively correlated as it gets further (negative parabola). However more importantly, the figure Latitude vs. Temperature illustrates the highest temperatures located at latitudes is between 0~30 degrees which includes the Tropic of Cancer (23.5 degrees). Although it may be assumed that cities near the equator have the highest temperatures, this phenomenon may be explained by the data's date (July) which is summer for the northern hemisphere thus yielding higher temperatures in latitudes of the northern hemisphere than those of the southern hemisphere. Also further research shows that temperatures are tend to be greater near the Tropic of Cancer and Capricorn (23.5 degrees north/south) than the equator (which also explains the reason for the majority of the Earth's deserts falling within these regions).

2. For the relationship between humidity and latitude, as well as the wind speed and latitude, there seems to be no significant relationship according to the figures. Relationships between latitude vs. humidity and wind speed may require time series analysis to see significant differences or patterns in respect to the latitude rather than a one day snapshot of the humidity and wind speed. 

3. The relationship between Cloudiness and Latitude also does not seem to be significantly illustrating anything worth noting, however, it does seem like slender lines are grouped around 20% intervals ranging from 0% - 100%, which may be explained by how the Cloudiness is quantified or whether or not this metric is generalized by such intervals. This relationship may also be suggested to be observed through a time series analysis to see more significant trends in relations to latitutde.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from datetime import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(cities)
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# initialize API call and counter variables
print("Begin Data Retrieval from OpenWeatherMap.org API")
cities_info = []
cities_printlog = []
city_count = 1
city_group = 1
city_not_found_count = 0

# obtaining information of cities through requests.get
for city in cities: 
    # set url and GET city info (JSON) from API
    city_url = (f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={api_key}&q={city}")
    city_info = requests.get(city_url).json()
    
    # print city log output progress and append to cities_printlog with city url
    log_output = f"Processing City {city_count} of Set {city_group} | {city}"
    print(log_output) 
    cities_printlog.append(log_output)
    cities_printlog.append(city_url)
    
    # append cities information if found otherwise output that city was not found
    if (city_info ["cod"] == ("404")): #adding exception for when city is not found
        print("CITY NOT FOUND.. Skip to next city")
        city_not_found_count +=1
    else:
        cities_info.append(city_info) #append new city when city found
    
    # city and city group counter 
    city_count += 1
    if (city_count == 50):
        city_count = 1
        city_group +=1

# print summary of the all cities' data from city list has been retrieved or not retrieved
print("=========================================================")
print(f"Data Retrieval Complete of {len(cities)-city_not_found_count} cities of {len(cities)} queries")
print(f"Cities not found: {city_not_found_count}")
print("=========================================================")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame


```python
# initialize variables
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
set_date = datetime.utcfromtimestamp(int(cities_info[0]['dt'])).strftime('%Y-%m-%d')

# organize city information (name, cloudiness, country, humidity, lat, lng, temp_max, and wind speed)in all retrieved cities
for city in cities_info:
    city_name.append(city['name'])
    cloudiness.append(city['clouds']['all'])
    country.append(city['sys']['country'])
    date.append(datetime.utcfromtimestamp(int(city['dt'])).strftime('%Y-%m-%d'))
    humidity.append(city['main']['humidity'])
    lat.append(city['coord']['lat'])
    lng.append(city['coord']['lon'])
    max_temp.append(city['main']['temp_max'])
    wind_speed.append(city['wind']['speed'])

# create df of cities' information
cities_info_df = {
            "City": city_name,
            "Cloudiness": cloudiness,
            "Country": country,
            "Date": date,
            "Humidity": humidity,
            "Lat": lat,
            "Lng": lng,
            "Temp max": max_temp,
            "Wind Speed": wind_speed
}
cities_info_df = pd.DataFrame(cities_info_df)

#export cities' information to CSV and print cities' information df
cities_info_df.to_csv("cities_data.csv")
cities_info_df
```

### Plotting the Data

```python
# create latitude vs. temperature plot with proper labels
plt.title(f"Latitude vs. Temperature Plot ({set_date})")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.scatter(cities_info_df["Lat"], cities_info_df["Temp max"])

#saving figure and show
plt.savefig("lat_vs_temp.png")
plt.show()
```
